```python
#luas_lingkaran=pi*r*r
import math
def hitung_luas_lingkaran(r=0):
  return math.pi * r**2
```

```python
luas1=hitung_luas_lingkaran(5)
luas2=hitung_luas_lingkaran(10)
print(f"Luas lingkaran dengan jari-jari=5 adalah {luas1:,.2f}")
print(f"Luas lingkaran dengan jari-jari=10 adalah {luas2:,.2f}")
print(f"Total Luas lingkaran  adalah {luas1+luas2:,.2f}")
```

```python
luas3=hitung_luas_lingkaran()
print(f"Luas lingkaran  adalah {luas3:,.2f}")

luas4=hitung_luas_lingkaran(100)
print(f"Luas lingkaran  adalah {luas4:,.2f}")
```

```python
def jumlah(a,b):
  print(f"{a}+{b}={a+b}")

jumlah(3,5)
#jumlah(3,5,6,7) akan error, karena jumlah parameter tidak sesuai
```

```python
def jumlahx(*bil):
  sum=0
  for i in bil:
    sum +=i

  print(f"Jumlah {bil} adalah {sum}")

jumlahx(21,32,44)
```

```python
jumlahx(2,2,2,5,6,7,8)
```

```python
def profile_user(**profile):
  for key,val in profile.items():
    print(f"{key} : {val}")
```

```python
profile_user(nama="Aktsa Alfajri", umur=21, email="aktsa@gmail.com", alamat="Cikarang")
profile_user(nama="Bambang", umur=20, alamat="Bogor")
```
