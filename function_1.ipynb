```python
def jam_aktual(jam_lembur):
  return 1.5+((jam_lembur-1)*2)

def uang_lembur(j_lembur,gapok):
  return (jam_aktual(j_lembur)*gapok)/173
```

```python
print(f"Uang lembur anda adalah  {uang_lembur(7,5000000):,.2f} rupiah")
```

```python
import math

def d(a,b,c):
  return b**2-4*a*c

def x1(a,b,d):
  return (-b+math.sqrt(d))/(2*a)

def x2(a,b,d):
  return (-b-math.sqrt(d))/(2*a)
```

```python
a,b,c=1,1,-6
det=d(a,b,c)

print(f"x1={x1(a,b,det)}")
print(f"x1={x2(a,b,det)}")
```
