## Global variabel

```python
name='Bambang'
def test():
  print(f'Nama {name}, diakses di dalam fungsi')

print(f'Nama {name}, diakses di luar fungsi')
test()
```

## Global variabel tidak bisa diubah di dalam fungsi:

```python
name='Bambang'
def test():
  name='Azka'
  print(f'Nama {name}, diakses di dalam fungsi')

test()
print(f'Nama {name}, diakses di luar fungsi')
```

```python
#Supaya bisa merubah di dalam fungsi, gunakan keyword global
name='Bambang'
def test():
  global name
  name='Azka'
  print(f'Nama {name}, diakses di dalam fungsi')

test()
print(f'Nama {name}, diakses di luar fungsi')
```

## Parameter dinamis list (*)

```python
#contoh list
bil=[11,23,31,40]
hasil=0
for i in bil:
  hasil=hasil+i

print(hasil)
```

```python
def jumlah(a,b): #ini parameter statis
  return a+b

def jumlahx(*bil): #ini parameter dinamis
  hasil=0
  for i in bil:
    hasil=hasil+i

  return hasil

```

```python
x=jumlah(34,12)
print(x)
```

```python
x=jumlah(1,2,5) # Errorn karena parameter statis
print(x)
```

```python
x=jumlahx(34,12)
print(x)
```

```python
x=jumlahx(1,2,5) #  karena parameter dinamis
print(x)
```

```python
nilai=jumlahx(100,86,90,76,55,40) #  karena parameter dinamis
print(nilai)
```

## Parameter dinamis Dictionary (**)

```python
def profile(name,email,address='Cikarang', age=20): # parameter statis
  print("Profile User")
  print(f"Nama {name}, alamat {address}, email {email} umur {age}")

profile(age=21,email='azka@gmail.com', name='Azka')
```

```python
profile(age=21,email='azka@gmail.com', name='Azka', pekerjaan='Mahasiswa')
```

```python
def profilex(**profile): # parameter statis
  print("Profile User")
  for key,val in profile.items():
    print(f"{key} : {val}")
```

```python
profilex(age=21,email='azka@gmail.com', name='Azka', pekerjaan='Mahasiswa')
```

```python
profilex(name='Azka',age=21,email='azka@gmail.com',status='Belum menikah', pekerjaan='Mahasiswa')
```
